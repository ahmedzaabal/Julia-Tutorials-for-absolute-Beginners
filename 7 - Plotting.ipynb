# using the package of plotting

````julia
using Plots
````

one of the advantages to plots.jl is that it allows you to seamlessly change backends.
<br>
in this notebook, we'll try out the ```gr()``` backend.

````julia
x = -3:0.1:3
f(x) = x ^ 2
y = f.(x)
````

### Let's load the GR() backend 

````julia
gr()
````

````julia
plot(x, y, label = "line")
scatter!(x,y,label = "points")
````

## Getting slightly fancier
<br>
now we will use scientific data and let's examine the relationship between the global tempreture and the number of pirates

````julia

global_temp=[14.4, 14.5,14.8,15.2,15.5,15.8]
num_of_pirates = [45000, 20000, 15000, 5000, 400, 17]

````

````julia
# First, PLot the data
plot(num_of_pirates, global_temp, legend = false)
scatter!(num_of_pirates, global_temp, legend = false)

#then flip the x axis we can see the tempreture changes as we move
xflip!()

# Adding titles and labels
xlabel!("Number of Pirates")
ylabel!("Temp")
title!("Influence of pirate population on global warming")
````

now, lets see how can we make subplots

````julia
p1 = plot(x,x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1,p2,p3,p4, layout=(2,2), legend=false)
````
